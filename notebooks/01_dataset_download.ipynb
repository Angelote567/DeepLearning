```python
from roboflow import Roboflow
import os, shutil

API_KEY = os.environ.get('ROBOFLOW_API_KEY') or 'PEGA_TU_API_KEY_AQUI'
rf = Roboflow(api_key=API_KEY)

# Dataset público qc/volleyball-hwxp2 v2 — 330 imágenes Tokio 2020, 3 clases (ball, player, referee)
project = rf.workspace('qc').project('volleyball-hwxp2')
version = project.version(2)

target = f'{DRIVE_ROOT}/data/volleyball-2'
if not os.path.exists(target):
    dataset = version.download('yolov8')   # descarga al cwd
    shutil.move(dataset.location, target)
    print('Dataset movido a', target)
else:
    print('Dataset ya existe en', target, '— saltando descarga')

# Symlink local para que el código del repo lo vea como data/volleyball-2/
os.makedirs('data', exist_ok=True)
if not os.path.exists('data/volleyball-2'):
    os.symlink(target, 'data/volleyball-2')
```

```python
import os
for split in ['train', 'valid', 'test']:
    p = f'{DRIVE_ROOT}/data/volleyball-2/{split}/images'
    if os.path.exists(p):
        n = len(os.listdir(p))
        print(f'{split}: {n} imágenes')
    else:
        print(f'{split}: (sin carpeta — algunos datasets solo tienen train/valid)')
```
