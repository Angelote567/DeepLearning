```python
from roboflow import Roboflow
import os

API_KEY = os.environ.get('ROBOFLOW_API_KEY') or 'PEGA_TU_API_KEY_AQUI'
rf = Roboflow(api_key=API_KEY)
project = rf.workspace().project('voley')
dataset = project.version(1).download('yolov8', location=f'{DRIVE_ROOT}/data/voley.yolov8')
print('Dataset descargado en', dataset.location)
```

```python
import os
for split in ['train', 'valid', 'test']:
    n = len(os.listdir(f'{DRIVE_ROOT}/data/voley.yolov8/{split}/images'))
    print(f'{split}: {n} imágenes')
```
